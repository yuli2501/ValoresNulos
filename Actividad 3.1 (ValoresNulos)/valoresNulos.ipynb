```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga desde un archivo.csv sin indice
data = pd.read_csv('Ventas_totales.csv')
data
```

```python
# Identificar valores nulos por columna
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
# Identificar valores nulos por dataframe
valores_nulos = data.isnull().sum().sum()
valores_nulos
```

```python
# Realizamos una copia del dataframe
data1 = data.copy()
```

MÉTODOS DE SUSTITUCIÓN DE VALORES NULOS 

Primer método de sustitución de valores nulos

Sustituir valores nulos con promedio o media

```python
data1["salon_ventas"] = data1["salon_ventas"].fillna(round(data["salon_ventas"].mean(),1))
data1["subtotal_ventas_alimentos_bebidas"] = data1["subtotal_ventas_alimentos_bebidas"].fillna(round(data["subtotal_ventas_alimentos_bebidas"].mean(),1))
data1["otros"] = data1["otros"].fillna(round(data["otros"].mean(),1))

# Identificar valores nulos por columna
valores_nulos = data1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data1.to_csv("Metodo1.csv")
```

Segundo método de sustitución de valores nulo

Sustituir valores nulos con mediana

```python
data1["tarjetas_debito"] = data1["tarjetas_debito"].fillna(round(data["tarjetas_debito"].median(),1))
data1["tarjetas_credito"] = data1["tarjetas_credito"].fillna(round(data["tarjetas_credito"].median(),1))
data1["otros_medios"] = data1["otros_medios"].fillna(round(data["otros_medios"].median(),1))

# Identificar valores nulos por columna
valores_nulos = data1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data1.to_csv("Metodo2.csv")
```

Tercer método de sustitución de valores nulos 

Sustituir valores nulos por un valor númerico en concreto

```python
data1["panaderia"] = data1["panaderia"].fillna(1)
data1["carnes"] = data1["carnes"].fillna(1)
data1["verduleria_fruteria"] = data1["verduleria_fruteria"].fillna(1)

# Identificar valores nulos por columna
valores_nulos = data1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data1.to_csv("Metodo3.csv")
```

Cuarto método de sustitución de valores nulos 

Sustituir valores nulos por un string en concreto

```python
data1["indumentaria_calzado_textiles_hogar"] = data1["indumentaria_calzado_textiles_hogar"].fillna("111111")
data1["electronicos_articulos_hogar"] = data1["electronicos_articulos_hogar"].fillna("111111")

# Corroboramos datos nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data1.to_csv("Metodo4.csv")
```

Quinto método de sustitución de valores nulos

Sustituir valores nulos por valores no nulos hacia adelante "forward fill"

```python
# Filtro por columnas
data1["bebidas"] = data1["bebidas"].fillna(method = "ffill")
data1["almacen"] = data1["almacen"].fillna(method = "ffill")

# Corroboramos datos nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data1.to_csv("Metodo5.csv")
```

Sexto método de sustitución de valores nulos 

Sustituir valores nulos no nulos hacia atrás backward fill ("bfill")

```python
# Filtro por columnas
data1["lacteos"] = data1["lacteos"].fillna(method = "bfill")
data1["alimentos_preparados_rotiseria"] = data1["alimentos_preparados_rotiseria"].fillna(method = "bfill")

# Corroboramos datos nulos
valores_nulos = data1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data1.to_csv("Metodo6.csv")
```
