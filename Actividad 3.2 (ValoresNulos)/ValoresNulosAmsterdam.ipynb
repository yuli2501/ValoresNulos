```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga desde un archivo.csv sin indice
data = pd.read_csv('Amsterdam.csv')
data
```

```python
columnas_a_conservar = [
    "id",
    "scrape_id",
    "last_scraped",
    "name",
    "host_id",
    "host_name",
    "host_since",
    "host_response_time",
    "host_response_rate",
    "host_acceptance_rate",
    "host_is_superhost",
    "host_listings_count",
    "host_total_listings_count",
    "host_identity_verified",
    "neighbourhood",
    "latitude",
    "longitude",
    "property_type",
    "room_type",
    "accommodates",
    "bathrooms",
    "bedrooms",
    "beds",
    "amenities",
    "price",
    "minimum_nights",
    "maximum_nights",
    "minimum_nights_avg_ntm",
    "maximum_nights_avg_ntm",
    "calendar_updated",
    "has_availability",
    "availability_30",
    "availability_90",
    "availability_365",
    "number_of_reviews",
    "first_review",
    "last_review",
    "review_scores_rating",
    "review_scores_accuracy",
    "review_scores_cleanliness",
    "review_scores_checkin",
    "review_scores_communication",
    "review_scores_location",
    "review_scores_value",
    "license",
    "instant_bookable",
    "calculated_host_listings_count",
    "calculated_host_listings_count_entire_homes",
    "calculated_host_listings_count_private_rooms",
    "calculated_host_listings_count_shared_rooms"
]

# Filtrar el DataFrame para conservar solo las columnas especificadas
data_filtrado = data[columnas_a_conservar]
```

```python
data_filtrado.to_csv("AmsterdamFiltrado.csv")
```

```python
# Identificar valores nulos por columna
valores_nulos = data_filtrado.isnull().sum()
valores_nulos
```

```python
# Identificar valores nulos por dataframe
valores_nulos = data_filtrado.isnull().sum().sum()
valores_nulos
```

```python
# Realizamos una copia del dataframe
data_filtrado1 = data_filtrado.copy()
```

COLUMNA 1 : host_response_time

Cuarto método de sustitución de valores nulos

Sustituir valores nulos por un string "undefined time" para decir que el anfitrión aún no tiene un tiempo definido para responder a las consultas de los huéspedes.

```python
data_filtrado1["host_response_time"] = data_filtrado1["host_response_time"].fillna("undefined time")

# Corroboramos datos nulos
valores_nulos=data_filtrado1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data_filtrado1.to_csv("Columna1.csv")
```

COLUMNA 2 : host_response_rate

Cuarto método de sustitución de valores nulos

Sustituir valores nulos por un string "0 %" para decir que el anfitrión aún no tiene un porcentaje definido de consultas a las que el anfitrión responde.

```python
data_filtrado1["host_response_rate"] = data_filtrado1["host_response_rate"].fillna("0%")

# Corroboramos datos nulos
valores_nulos=data_filtrado1.isnull().sum()
valores_nulos 
```

```python
# Convertir dataframe a csv
data_filtrado1.to_csv("Columna2.csv")
```

COLUMNA 3 : host_acceptance_rate 

Cuarto método de sustitución de valores nulos

Sustituir valores nulos por un string "0 %" para decir que el anfitrión aún no tiene un porcentaje definido de solicitudes de reserva que el anfitrión acepta.

```python
data_filtrado1["host_acceptance_rate"] = data_filtrado1["host_acceptance_rate"].fillna("0%")

# Corroboramos datos nulos
valores_nulos=data_filtrado1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data_filtrado1.to_csv("Columna3.csv")
```

COLUMNA 4 : host_is_superhost

Quinto método de sustitución de valores nulos

Sustituir valores nulos por valores no nulos hacia adelante "forward fill", para indicar si el anfitrión tiene el estatus de “Superanfitrión”

```python
data_filtrado1["host_is_superhost"] = data_filtrado1["host_is_superhost"].fillna(method = "ffill")

# Corroboramos datos nulos
valores_nulos=data_filtrado1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data_filtrado1.to_csv("Columna4.csv")
```

COLUMNA 5 : neighbourhood 

Cuarto método de sustitución de valores nulos

Sustituir valores nulos por un string "No definido" para decir que aún no tiene un nombre definido del vecindario.

```python
data_filtrado1["neighbourhood"] = data_filtrado1["neighbourhood"].fillna("undefined")

# Corroboramos datos nulos
valores_nulos=data_filtrado1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data_filtrado1.to_csv("Columna5.csv")
```

COLUMNA 6 : bathrooms 

Tercer método de sustitución de valores nulos 

Sustituir valores nulos por un valor númerico en concreto, en este caso seria "1" porque todos los espacios de renta deberían de tener como mínimo un baño. 

```python
data_filtrado1["bathrooms"] = data_filtrado1["bathrooms"].fillna(1)

# Identificar valores nulos por columna
valores_nulos = data_filtrado1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data_filtrado1.to_csv("Columna6.csv")
```

COLUMNA 7 : bedrooms

Segundo método de sustitución de valores nulo

Sustituir valores nulos con mediana, para sacar un número medio que indique cuantos dormitorios puede tener ese lugar de renta

```python
data_filtrado1["bedrooms"] = data_filtrado1["bedrooms"].fillna(round(data["bedrooms"].median(),1))

# Identificar valores nulos por columna
valores_nulos = data_filtrado1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data_filtrado1.to_csv("Columna7.csv")
```

COLUMNA 8 : beds

Segundo método de sustitución de valores nulo

Sustituir valores nulos con mediana, para sacar un número medio que indique cuantas camas puede tener ese lugar de renta

```python
data_filtrado1["beds"] = data_filtrado1["beds"].fillna(round(data["beds"].median(),1))

# Identificar valores nulos por columna
valores_nulos = data_filtrado1.isnull().sum()
valores_nulos
```

```python
# Convertir dataframe a csv
data_filtrado1.to_csv("Columna8.csv")
```
